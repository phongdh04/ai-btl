```python
# xoa du lieu trung (so luon ti le bao nhieu so voi du lieu mau)
data = data.drop_duplicates()
```

```python
def binarization(i):
    if i >= 7:
      return 1
    else:
      return 0
```

```python
X = data.drop(['quality'], axis=1)
y = data['quality']
```

```python
y_try = y.apply(binarization)
```

Rượu vang chất lượng cao (1) cho thấy nồng độ cồn trung bình cao hơn so với rượu vang chất lượng thấp (0). Bằng cách xem xét điều này, nồng độ cồn có thể là một dự đoán tốt về chất lượng vì có nồng độ cồn cao hơn cho thấy có chất lượng cao hơn.

Rượu vang chất lượng cao có phạm vi liên tứ phân vị (IQR) rộng hơn, có nghĩa là rượu vang chất lượng cao có sự khác biệt lớn hơn về nồng độ cồn so với rượu vang chất lượng thấp.

Có một số ngoại lệ về nồng độ cồn đối với rượu vang chất lượng thấp, vì thông thường nồng độ cồn cao hơn khiến rượu vang có chất lượng cao hơn, chúng ta có thể nói rằng đây là một tình huống bất thường. Bởi vì mặc dù chúng có nồng độ cồn cao nhưng chúng có chất lượng thấp.

```python
# gọi các thư viện cần thiết
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_curve, roc_auc_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
import pydotplus
from IPython.display import Image
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_try, test_size=0.2, random_state=42)

# xây dựng model
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(X_train, y_train)

# dự đoán với tập test
y_pred = dtc.predict(X_test)

# điểm accuracy
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)
```
